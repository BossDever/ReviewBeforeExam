```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/iris.csv')
print("1. CSV file has been read:")
print(df.head())
```

```python
features = df.columns[:-1]
means = df[features].mean()
stds = df[features].std()
variances = df[features].var()
print("\\n2. Calculated the mean, standard deviation, and variance:")
print("Means:\\n", means)
print("Standard Deviations:\\n", stds)
print("Variances:\\n", variances)
```

```python
df_normalized = (df[features] - means) / stds
print("\\n3. Data has been normalized:")
print(df_normalized.head())
```

```python
plt.figure(figsize=(10, 6))
species = df['virginica'].unique()
for spec in species:
    plt.scatter(df_normalized[df['virginica'] == spec]['setosa'],
                df_normalized[df['virginica'] == spec]['versicolor'],
                label=spec)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
print("4. Scatter chart has been created.")
```
